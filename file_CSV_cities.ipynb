## Работа с данными о городах: CSV, Python

```python
import csv
```

```python
file_cities_csv = "cities.csv"
```

```python
with open(file_cities_csv) as f:
    reader = csv.DictReader(f, delimiter=';')
    print(reader.fieldnames)
```

```python
with open(file_cities_csv) as f:
    reader = csv.DictReader(f, delimiter=';')
    for line in reader:
        print(line['Город'])
```

```python
# вывести в консоль список всех регионов (только уникальные значения) и отсортировать по алфавиту
regions = []
with open(file_cities_csv) as f:
    readerDict = csv.DictReader(f, delimiter=';')
    for line in readerDict:
        regions.append(line['Регион'])
print(sorted(set(regions)))
```

```python
# вывести в консоль список городов, в названии которых есть буква '-'
cities = []
with open(file_cities_csv) as f:
    readerDict = csv.DictReader(f, delimiter=';')
    for line in readerDict:
        if '-' in line['Город']:
            cities.append(line['Город'])

print(cities)
```

```python
# создать csv-файл, где будут храниться данные о городах, в названии которых есть буква 'ы'
temp_data2 = []
with open(file_cities_csv) as f:
    reader = csv.DictReader(f, delimiter=';')
    temp_header = reader.fieldnames
    for line in reader:
        if "ы" in line["Город"]:
            temp_data2.append(line) 

with open("temp_cities_bukv_y.csv", 'w') as f:
    writer = csv.DictWriter(f, lineterminator='\n', delimiter=';', fieldnames=temp_header)
    writer.writeheader()
    writer.writerows(temp_data2)
```

```python
# создать csv-файл, где данные будут отсортированы по названию округа, региона, города
temp_data3 = []
with open(file_cities_csv) as f:
    reader = csv.DictReader(f, delimiter=';')
    temp_header = reader.fieldnames
    for line in reader:
        temp_data3.append(line)
        
with open("temp_cities_sort.csv", 'w') as f:
    writer = csv.DictWriter(f, lineterminator='\n', delimiter=';', fieldnames=temp_header)
    writer.writeheader()
    sorted_temp = sorted(temp_data3, key = lambda x: (x['Федеральный округ'], x['Регион'], x['Город']))
    writer.writerows(sorted_temp)
```

```python
# вывести в консоль список городов, название которых содержит до 4-х букв включительно и отсортированные в обратном алфавитном порядке
cities = []
with open(file_cities_csv) as f:
    readerDict = csv.DictReader(f, delimiter=';')
    for line in readerDict:
        if len(line['Город']) <= 4:
            cities.append(line['Город'])

print(sorted(cities, reverse=True))
```

```python
# вывести в консоль список регионов и количество городов в каждом регионе
cities = {}
with open(file_cities_csv) as f:
    readerDict = csv.DictReader(f, delimiter=';')
    for line in readerDict:
        if line['Регион'] not in cities:
            cities[line['Регион']] = 1
        else:
            cities[line['Регион']] += 1

print(list(cities.items()))
```

```python
# вывести в консоль буквы алфавита по убыванию и рядом количество городов, начинающихся на каждую букву
alphabet = {}
with open(file_cities_csv) as f:
    readerDict = csv.DictReader(f, delimiter=';')
    for line in readerDict:
        for i in range(33):
            if chr(1040+i) not in alphabet:
                alphabet[chr(1040+i)] = 0
            elif chr(1040+i) == line['Город'][0]:
                alphabet[chr(1040+i)] += 1
print(alphabet)
```

```python
# вывести в консоль числа по возрастанию и рядом количество городов, состоящих из указанного количества букв
len_cities = {}
with open(file_cities_csv) as f:
    readerDict = csv.DictReader(f, delimiter=';')
    for line in readerDict:
        for i in range(30):
            if i not in len_cities:
                len_cities[i] = 0
            if i == len(line['Город']):
                len_cities[i] += 1
print(len_cities)
```

```python
# вывести в консоль список ТОП-10 городов с самыми длинными названиями
len_cities = []
with open(file_cities_csv) as f:
    readerDict = csv.DictReader(f, delimiter=';')
    for line in readerDict:
        len_cities.append(line['Город'])

len_cities_new = sorted(len_cities, key=len, reverse=True)
print(len_cities_new[:10])
```

```python
# вывести в консоль список ТОП-10 городов, в названии которых больше всего согласных букв
sogl_cities = {}
sogl = ['б', 'Б', 'в', 'В', 'г', 'Г', 'д', 'Д', 'ж', 'Ж', 'з', 'З', 'й', 'Й', 'к', 'К', 'л', 'Л', 'м', 'М', 'н', 'Н', 'п', 'П', 'р', 'Р', 'с', 'С', 'т', 'Т', 'ф', 'Ф', 'х', 'Х', 'ц', 'Ц', 'ч', 'Ч', 'ш', 'Ш', 'щ', 'Щ']
with open(file_cities_csv) as f:
    readerDict = csv.DictReader(f, delimiter=';')
    for line in readerDict:
        sogl_cities[line['Город']] = 0
        for i in range(len(sogl)):
            sogl_cities[line['Город']] += line['Город'].count(sogl[i])
        
sogl_cities_new = sorted(sogl_cities.items(), key=lambda item: item[1], reverse=True)
print(sogl_cities_new[:10])
```

```python
# вывести в консоль список ТОП-10 букв, которые используются чаще всего в названии городов
bukvi_cities = {}
bukvi = ['a', 'А', 'б', 'Б', 'в', 'В', 'г', 'Г', 'д', 'Д', 'е', 'Е', 'ё', 'Ё', 'ж', 'Ж', 'з', 'З', 'й', 'Й', 'к', 'К', 'л', 'Л', 'м', 'М', 'н', 'Н', 'о', 'О', 'п', 'П', 'р', 'Р', 'с', 'С', 'т', 'Т', 'у', 'У', 'ф', 'Ф', 'х', 'Х', 'ц', 'Ц', 'ч', 'Ч', 'ш', 'Ш', 'щ', 'Щ', 'ъ', 'ы', 'ь', 'э', 'Э', 'ю', 'Ю', 'я', 'Я']
with open(file_cities_csv) as f:
    readerDict = csv.DictReader(f, delimiter=';')
    for line in readerDict:
        for i in range(len(bukvi)):
            if bukvi[i] in bukvi_cities:
                bukvi_cities[bukvi[i]] += line['Город'].count(bukvi[i])
            else:
                bukvi_cities[bukvi[i]] = line['Город'].count(bukvi[i])

bukvi_cities_new = sorted(bukvi_cities.items(), key=lambda item: item[1], reverse=True)
print(bukvi_cities_new[:10])
```

```python
import math

def distance(lat1, lon1, lat2, lon2):
    R = 6371  # радиус Земли в километрах
    dlat = math.radians(lat2 - lat1)
    dlon = math.radians(lon2 - lon1)
    P = math.sin(dlat/2) ** 2 + math.cos(math.radians(lat1)) * math.cos(math.radians(lat2)) * math.sin(dlon/2) ** 2
    Q = 2 * math.atan2(math.sqrt(P), math.sqrt(1 - P))
    return(Q * R)
```

```python
# вывести в консоль расстояние между случайными двумя городами (используем данные из lat и lng)
from random import randint

cities = {}
with open(file_cities_csv) as f:
    readerDict = csv.DictReader(f, delimiter=';')
    for line in readerDict:
        cities[line['Город']] = [line['lat'].replace(',', '.'), line['lng'].replace(',', '.')]

data = list(cities.items())
key, value = data[randint(0, len(data))]
key2, value2 = data[randint(0, len(data))]
print(key, value)
print(key2, value2)
print(round(distance(float(value2[0]), float(value2[1]), float(value[0]), float(value[1])), 3))
```

```python
# вывести в консоль список городов и расстоянием до города 'Москва' и отсортировать города по удаленности от Москвы

cities = {}
with open(file_cities_csv) as f:
    readerDict = csv.DictReader(f, delimiter=';')
    for line in readerDict:
        if line['Город'] == 'Москва':
            value1_moscow = line['lat'].replace(',', '.')
            value2_moscow = line['lng'].replace(',', '.')
        else:
            cities[line['Город']] = [line['lat'].replace(',', '.'), line['lng'].replace(',', '.')]

cities_new = []
for key in cities:
    values = cities[key]
    cities_new.append([key, round(distance(float(value1_moscow), float(value2_moscow), float(values[0]), float(values[1])), 3)])

print(sorted(cities_new, key=lambda x: x[1]))
```

```python
# вывести в консоль список ТОП-10 городов, которые дальше всего расположены друг от друга

with open(file_cities_csv) as f:
    readerDict = csv.DictReader(f, delimiter=';')
    cities = []
    for line in readerDict:
        cities.append({
            'name': line['Город'],
            'latitude': float(line['lat'].replace(',', '.')),
            'longitude': float(line['lng'].replace(',', '.'))
        })

distances = {}
for i in range(len(cities)):
    for j in range(i+1, len(cities)):
        distances[(i, j)] = distance(cities[i]['latitude'], cities[i]['longitude'], cities[j]['latitude'], cities[j]['longitude'])

top10 = sorted(distances.items(), key=lambda x: x[1], reverse=True)[:10]
for (i, j), dist in top10:
    print(f"{cities[i]['name']} : {cities[j]['name']} = {dist:.2f} км")
```
